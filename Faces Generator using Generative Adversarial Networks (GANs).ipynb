```python
from faces_generator import FacesGenerator

# This line creates an instance of the FacesGenerator, it will receive a directory 
# containing the faces to be trained.
generator = FacesGenerator(example_faces_dir='faces/')
```

```python
# Extract the faces found in the images of the example faces directory.
# This will create a dataset file called imgs_128.npz compressed to be used by the GANs.
generator.extract_faces(dataset_file_name='imgs_128.npz')
```

```python
# Plot the extracted faces
# This will show you the first 16 faces of your dataset.
generator.plot_faces(dataset_file_name='imgs_128.npz')
```

```python
# Generate Faces
# This will generate the faces, the faces will appear in your root directory in png
# and h5 format.
# plot_008x008-tuned.png
# model_008x008-tuned.h5
generator.generate_faces(dataset_file_name='imgs_128.npz')
```
